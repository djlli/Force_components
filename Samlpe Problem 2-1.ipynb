## Sample Problem 2/1 p. 29

Determine componets of the force $ F_1, F_2 $ and $ F_3 $

## Calculate F1

```python
%%render 1
alpha = radians(35) # Input in degrees
F_1 = 600 * si.N

F_1x = F_1 * cos(alpha)
F_1Y = F_1 * sin(alpha)
```

## Calculate F2

```python
%%render 0

ratio = 3/4
mod = (3 * si.m)
hos = (4 * si.m)
hyp = (mod**2 + hos**2)**0.5
F_2 = (500 * si.N)

F_2x = F_2 * hos/hyp
F_2y = F_2 * mod/hyp
```

## Calculate F3

```python
%%render 1

x1 = (0.2 * si.m)
x2 = (0.4 * si.m)

y1 = (0.1 * si.m)
y2 = (0.3 * si.m)

F_3 = 800 * si.N

Delta_x = x2-x1
Delta_y = y2+y1

beta = atan (Delta_y / Delta_x)

F_3x = F_3 * cos(beta)
F_3y = -F_3 * sin(beta)
```
